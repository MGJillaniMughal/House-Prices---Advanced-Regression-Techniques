# Now Imports Some Others Lib's

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
```

```python
num_por = Pipeline(steps = [
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('scaler', MinMaxScaler())
])
```

```python
cat_por = Pipeline(steps = [
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown = 'ignore', sparse = False))
])
```

```python
prepor=ColumnTransformer(
[
    ('num',num_por,n_cols),
    ('cat',cat_por,cati_cols)
]
)
```

# Now Import Models

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dec_tree_reg_mod=DecisionTreeRegressor(max_depth=17,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rand_for_reg=RandomForestRegressor(n_estimators=169,random_state=0)
```

```python
from xgboost import XGBRegressor
```

```python
xgb_reg=XGBRegressor(n_estimators=301,learing_rate=0.1,random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
clf=Pipeline(steps=[
                ('preprocessor',prepor),
                ('dec_tree_reg_mod',dec_tree_reg_mod)
            ])
```

```python
my_columns=n_cols+cati_cols
x=train_df[my_columns]
y=np.log1p(train_df[Y])
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
def Model(models):
    clf = Pipeline(steps = [
      ('preprocessor',prepor),
      ('model',models)
    ])
    clf.fit(X_train, y_train)
    print(models)
    print("Train set score:", clf.score(X_train, y_train))
    print("Test set score:", clf.score(X_test, y_test))
    print("Train set mse:", mean_squared_error(y_train, clf.predict(X_train)))
    print("Test set mse:", mean_squared_error(y_test, clf.predict(X_test)))
    print("\n")
```

# Now Check All Models

```python
Models=[dec_tree_reg_mod,rand_for_reg,xgb_reg]
for i in Models:
    Model(i)
```
